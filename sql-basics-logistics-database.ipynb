```python
import pandas as pd

routes = pd.read_csv('/kaggle/input/datasets/yogape/logistics-operations-database/routes.csv')
facilities = pd.read_csv('/kaggle/input/datasets/yogape/logistics-operations-database/facilities.csv')
drivers = pd.read_csv('/kaggle/input/datasets/yogape/logistics-operations-database/drivers.csv')
trips = pd.read_csv('/kaggle/input/datasets/yogape/logistics-operations-database/trips.csv')
safety_incidents = pd.read_csv('/kaggle/input/datasets/yogape/logistics-operations-database/safety_incidents.csv')
trailers = pd.read_csv('/kaggle/input/datasets/yogape/logistics-operations-database/trailers.csv')
delivery_events = pd.read_csv('/kaggle/input/datasets/yogape/logistics-operations-database/delivery_events.csv')
driver_monthly_metrics = pd.read_csv('/kaggle/input/datasets/yogape/logistics-operations-database/driver_monthly_metrics.csv')
trucks = pd.read_csv('/kaggle/input/datasets/yogape/logistics-operations-database/trucks.csv')
loads = pd.read_csv('/kaggle/input/datasets/yogape/logistics-operations-database/loads.csv')
fuel_purchases = pd.read_csv('/kaggle/input/datasets/yogape/logistics-operations-database/fuel_purchases.csv')
truck_utilization_metrics = pd.read_csv('/kaggle/input/datasets/yogape/logistics-operations-database/truck_utilization_metrics.csv')
customers = pd.read_csv('/kaggle/input/datasets/yogape/logistics-operations-database/customers.csv')
maintenance_records = pd.read_csv('/kaggle/input/datasets/yogape/logistics-operations-database/maintenance_records.csv')
```

```python
tables = {
    "routes": routes,
    "facilities": facilities,
    "drivers": drivers,
    "trips": trips,
    "safety_incidents": safety_incidents,
    "trailers": trailers,
    "delivery_events": delivery_events,
    "driver_monthly_metrics": driver_monthly_metrics,
    "trucks": trucks,
    "loads": loads,
    "fuel_purchases": fuel_purchases,
    "truck_utilization_metrics": truck_utilization_metrics,
    "customers": customers,
    "maintenance_records": maintenance_records
}

for name, df in tables.items():
    print(f"{name:<30}: {df.shape[0]} rows {df.shape[1]} columns")
```

```python
import sqlite3

conn = sqlite3.connect(':memory:')  # database di RAM

routes.to_sql('routes', conn, index=False, if_exists='replace')
facilities.to_sql('facilities', conn, index=False, if_exists='replace')
drivers.to_sql('drivers', conn, index=False, if_exists='replace')
trips.to_sql('trips', conn, index=False, if_exists='replace')
safety_incidents.to_sql('safety_incidents', conn, index=False, if_exists='replace')
trailers.to_sql('trailers', conn, index=False, if_exists='replace')
delivery_events.to_sql('delivery_events', conn, index=False, if_exists='replace')
driver_monthly_metrics.to_sql('driver_monthly_metrics', conn, index=False, if_exists='replace')
trucks.to_sql('trucks', conn, index=False, if_exists='replace')
loads.to_sql('loads', conn, index=False, if_exists='replace')
fuel_purchases.to_sql('fuel_purchases', conn, index=False, if_exists='replace')
truck_utilization_metrics.to_sql('truck_utilization_metrics', conn, index=False, if_exists='replace')
customers.to_sql('customers', conn, index=False, if_exists='replace')
maintenance_records.to_sql('maintenance_records', conn, index=False, if_exists='replace')
```
